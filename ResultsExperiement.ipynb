```python
import matplotlib.pyplot as plt
```

```python
file1 = "./Time_Based_Model.txt"
file2 = "./Agent_Based_Model.txt"

file3 = "./Time_Based_Model_Skewed.txt"
file4 = "./Agent_Based_Model_Skewed.txt"

file5 = "./Roundabout_Medium_Intesity.txt"
file6 = "./Roundabout_High_Intensity.txt"
```

```python
def read_file(file_name):
    content = []
    with open(file_name, "r") as f:
        content = [ i.split(",") for i in f.read().split("\n") ][:-1:]
        content = [ list(map(int, i)) for i in content ]
    return content
```

```python
file1_content = read_file(file1)
file2_content = read_file(file2)
file3_content = read_file(file3)
file4_content = read_file(file4)
file5_content = read_file(file5)
file6_content = read_file(file6)
x_axis = list(range(len(file1_content)))
```

```python
plt.figure( figsize=(20,10) )
plt.plot( x_axis , file1_content ,label="Time_Based_Model")
plt.plot( x_axis , file2_content ,label="Agent_Based_Model")
plt.plot( x_axis , file5_content ,label="RoundAbout_Model")
plt.legend()
plt.xlabel("Time Traversed")
plt.ylabel("total no. of cars waiting")
plt.show()
```

```python
file_3_0 = [ i[0] for i in file3_content ]
file_3_1 = [ i[1] for i in file3_content ]

file_4_0 = [ i[0] for i in file4_content ]
file_4_1 = [ i[1] for i in file4_content ]
```

```python
plt.figure( figsize=(20,10) )
plt.plot( x_axis , file_3_0 ,label="Time_Based_Model")
plt.plot( x_axis , file_4_0 ,label="Agent_Based_Model")
plt.plot( x_axis , file6_content ,label="RoundAbout_Model")
plt.legend()
plt.xlabel("Time Traversed")
plt.ylabel("total no. of cars waiting")
plt.show()
```
